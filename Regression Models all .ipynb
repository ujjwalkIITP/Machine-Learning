```python
import sklearn
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dataset = pd.read_csv('/content/FDataReg.csv')
```

```python
X = dataset.iloc[:,:-1]
Y = dataset.iloc[:,-1]
Y = np.array(Y)
Y = Y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression()
reg1.fit(X_train, Y_train)
```

```python
Y_pred = reg1.predict(X_test)
```

```python
print(Y_pred)
```

```python
print(Y_test)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test, Y_pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 4)
X_poly = poly.fit_transform(X_train)
reg2 = LinearRegression()
reg2.fit(X_poly, Y_train)
```

```python
Y_poly_pred = reg2.predict(poly.fit_transform(X_test))
```

```python
r2_score(Y_test, Y_poly_pred)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_Y = StandardScaler()
X_train_SVR = sc_X.fit_transform(X_train)
Y_train_SVR = sc_Y.fit_transform(Y_train)

```

```python
from sklearn.svm import SVR
regressor_SVR = SVR(kernel='rbf')
regressor_SVR.fit(X_train_SVR, Y_train_SVR)
```

```python
Y_pred_SVR = sc_Y.inverse_transform(regressor_SVR.predict(sc_X.transform(X_test)).reshape(-1,1))
```

```python
r2_score(Y_test, Y_pred_SVR)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_forest = RandomForestRegressor(n_estimators = 10, random_state= 0)
regressor_forest.fit(X_train, Y_train)
```

```python
Y_forest_pred = regressor_forest.predict(X_test)
```

```python
r2_score(Y_test, Y_forest_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor_decision = DecisionTreeRegressor(random_state = 0)
regressor_decision.fit(X_train, Y_train)
```

```python
Y_decision_predict = regressor_decision.predict(X_test)
```

```python
r2_score(Y_test, Y_decision_predict)
```
