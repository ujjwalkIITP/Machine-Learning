```python
import requests
import pandas as pd
import time
```

```python
API_Key = "sample"
channelID = "sample"
```

```python
pagetoken=""
url = "https://www.googleapis.com/youtube/v3/search?key="+API_Key+"&channelId="+channelID+"&part=snippet,id&order=date&maxResults=10000"+pagetoken
```

```python
response = requests.get(url).json()
```

```python
publish_date = response['items'][0]['snippet']['publishedAt']
publish_date=str(publish_date).split('T')
publish_date = publish_date[0]
publish_date
```

```python
Title = response['items'][0]['id']['videoId']
Title
```

```python
df2= pd.DataFrame(columns = ['ID','Title', 'Publish_date', 'View', 'Like', 'Comment','Favorite'])
```

```python
df2
```

```python
ID_arr = []
Title_arr = []
Publish_date_arr = []
View_count = []
Like_Count =[]
Comment_Count = []
Favorite_Count = []
```

```python

for video in response["items"]:
  if video['id']['kind']=='youtube#video':
 # ID = video['id']['videoId']
      ID = video['id']['videoId']
      Title = video['snippet']['title']
      publish_date = video['snippet']['publishedAt']
      publish_date=str(publish_date).split('T')
      publish_date = publish_date[0]

      url_video_stats= "https://www.googleapis.com/youtube/v3/videos?id="+ID+"&part=statistics&key="+API_Key
      response_video_stats = requests.get(url_video_stats).json()

      view_count = response_video_stats['items'][0]['statistics']['viewCount']

      like_count = response_video_stats['items'][0]['statistics']['likeCount']

      favorite_count = response_video_stats['items'][0]['statistics']['favoriteCount']

      comment_count = response_video_stats['items'][0]['statistics']['commentCount']

      df2 = df2.append({ 'ID': ID,
                      'Title': Title, 
                      'Publish_date': publish_date, 
                      'View': view_count, 
                      'Like': like_count, 
                      'Comment': comment_count,
                      'Favorite': favorite_count},ignore_index=True)


  

```
