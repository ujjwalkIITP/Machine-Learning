```python
import pandas as pd
from matplotlib import pyplot as plt
import sklearn 
```

```python
dataset = pd.read_csv('/content/Market_Basket_Optimisation_CSV.csv',header = None )
```

```python
dataset.shape
```

```python
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
transactions2 = []
for i in range(0,7501):
  transactions2.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
len(transactions2)
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2 )
```

```python
results = list(rules)
```
