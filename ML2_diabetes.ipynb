```python
import pandas as pd
import sklearn
import numpy as np
```

```python
dataset = pd.read_csv('/content/ML2_diabetes.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,:-1]
Y = dataset.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(X_train, Y_train)
```

```python
Y_test_pred = neigh.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score( Y_test, Y_test_pred)
```

```python
score
```

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
regressor.fit(X_train, Y_train)
```

```python
Y_test_pred2 = regressor.predict(X_test)
```

```python
score2 = accuracy_score( Y_test, Y_test_pred2)
```

```python
score2
```

```python
from sklearn.tree import DecisionTreeClassifier
tree1 = DecisionTreeClassifier()
tree1.fit(X_train, Y_train)
```

```python
Y_test_pred3 = tree1.predict(X_test)
```

```python
score3 = accuracy_score(Y_test, Y_test_pred3)
```

```python
score3
```

```python
from sklearn.ensemble import RandomForestClassifier
forest1 = RandomForestClassifier()
forest1.fit(X_train, Y_train)
```

```python
Y_test_pred4 = forest1.predict(X_test)
```

```python
score4 = accuracy_score( Y_test, Y_test_pred4)
```

```python
score4
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, Y_train)
```

```python
Y_test_pred5 = gnb.predict(X_test)
```

```python
score5 = accuracy_score( Y_test, Y_test_pred5,)
```

```python
score5
```
