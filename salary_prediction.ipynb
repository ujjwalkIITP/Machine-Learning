```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('/content/Salary_Data.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
X = dataset.iloc[:,:-1]
Y = dataset.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, Y)
```

```python
Y_pred = regressor.predict(X)
```

```python
from matplotlib import pyplot as plt
plt.plot(X, Y)
```

```python
import seaborn as sns
sns.scatterplot(x = "YearsExperience", y = "Salary", data = dataset, color = "Blue")
plt.plot(X, Y_pred, color = 'Red')

plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2)
x_poly = poly.fit_transform(X)
model2 = LinearRegression()
model2.fit(x_poly, Y)
```

```python
r_square = regressor.score(X, Y)
```

```python
r_square
```

```python
r_square2 = model2.score(x_poly, Y)
```

```python
r_square2
```

```python
from sklearn.ensemble import RandomForestRegressor
reg3 = RandomForestRegressor()
reg3.fit(X, Y)
```

```python
r_square4 = reg3.score(X, Y)
```

```python
r_square4
```

```python
plt.plot(X, reg3.predict(X), color = 'Red')
plt.scatter(X, Y, color = 'Blue')
plt.plot()
```

```python
from sklearn.tree import DecisionTreeRegressor
reg4 = DecisionTreeRegressor()
reg4.fit(X, Y)
```

```python
r_score5 = reg4.score(X, Y)
```

```python
r_score5
```

```python
plt.plot(X, reg4.predict(X), color = 'Red')
plt.scatter(X, Y, color = 'Blue')
plt.plot()
```
